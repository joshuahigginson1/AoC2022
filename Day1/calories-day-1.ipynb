```python
# Imports
import pandas as pd
  
```

```python
# Read text file into pandas DataFrame
df = pd.read_table("./input.txt", delimiter=" ", skip_blank_lines=False, header=None)

# Rename header.
df.rename(columns={0: 'Calories'}, inplace=True)

df.head()
```

```python
# Count down the Dataframe count for any occurences of NaN. Cumulatively sum as we go down the list. Add 1 to get the correct Elf ID.
df["Elf_ID"] = df.isnull().all(axis=1).cumsum() + 1

df.head(10)
```

```python
# Get rid of the NAN markers.
df_clean = df.dropna()

df_clean.head(20)
```

```python
# Group by Elf ID, and sum the calories.
df_aggregated = df_clean.groupby(["Elf_ID"]).sum()
df_aggregated.head()
```

```python
df_sorted = df_aggregated.sort_values(["Calories"], ascending=False)

top_1 = df_sorted["Calories"].nlargest(1).sum()

print(top_1)
```

```python
# Get the Top 3 elves' values and sum them together.

top_3 = df_sorted["Calories"].nlargest(3).sum()

print(top_3)
```
